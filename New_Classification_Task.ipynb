## Mushroom Classification Project

### Problem Description
The goal of this project is to classify mushrooms as **edible (e)** or **poisonous (p)** based on 20 descriptive features such as cap shape, cap color, stem characteristics, habitat, and season.  
This is a **binary classification problem** where the target variable is the mushroom class (`edible` or `poisonous`).  

---

### Project Workflow

#### 1. Exploratory Data Analysis (EDA)
- Inspect the dataset structure (rows, columns, data types).  
- Check for missing values and duplicates.  
- Visualize distributions of categorical features (bar plots) and numeric features (histograms, boxplots).  
- Explore correlations between features and the target class.  
- Look for class imbalance in the target variable (edible vs poisonous).  

#### 2. Data Cleaning & Preprocessing
- Handle missing values or unknown entries if present.  
- Encode categorical features (e.g., **One-Hot Encoding** or **Label Encoding**).  
- Scale numerical features (e.g., **StandardScaler** or **MinMaxScaler**).  
- Stratify the dataset when splitting into **training and test sets** to maintain class balance.  
- If class imbalance is significant, apply oversampling techniques (e.g., **SMOTE**).  

#### 3. Feature Selection
To reduce dimensionality and improve generalization, apply feature selection techniques such as:  
- **Filter Methods:** Use statistical tests (e.g., Chi-square, ANOVA F-test, mutual information) to rank features.  
- **Wrapper Methods:** Apply Recursive Feature Elimination (**RFE**) with models like Logistic Regression or Decision Trees.  
- **Embedded Methods:** Leverage feature importance from models (e.g., Decision Tree, Random Forest, Lasso Regression).  
- Compare model performance before and after feature selection to evaluate impact.

Train and evaluate the following classification models:  
- **Logistic Regression**

- **Support Vector Machine (SVM)**

For each model:  
- Train on the **training set**.

  - **Classification Report** (Precision, Recall, F1-score, Accuracy)

#### 5. Hyperparameter Tuning
To improve model performance, apply hyperparameter optimization techniques:  
- **Grid Search:** Exhaustively search all combinations of parameters (good for small parameter spaces).  
- **Random Search:** Randomly sample parameter combinations (more efficient for large parameter spaces).  
- **Bayesian Optimization:** Iteratively choose hyperparameters based on past performance (efficient for complex models like Random Forest, SVM, or KNN).  

Compare tuned models with baseline models to measure improvement.  

#### 6. Feature Importance
- Extract **feature importance** from Decision Tree and Random Forest models.  
- Rank features to identify the most influential ones for classification.  
- Use feature importance as a basis for feature selection and model simplification.  

#### 7. Model Comparison

```python
# Step 2: Data Preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
# Identify numerical and categorical features
numeric_features = ['cap-diameter', 'stem-height', 'stem-width']
categorical_features = X.select_dtypes(include=['object']).columns.tolist()
```

```python
X = X.drop_duplicates().reset_index(drop=True) # Remove duplicate rows
```

```python
X.drop(columns=['veil-color','spore-print-color','veil-type','stem-root'], inplace=True) # Drop columns with high cardinality or irrelevant features
```

```python
unknown_tokens = ['?', 'unknown', 'None', 'NA', 'N/A', 'na', 'nan',' ']
X = X.replace(unknown_tokens, np.nan)
```

```python
from sklearn.impute import SimpleImputer
# For numerical features, impute with median
numerical_imputer = SimpleImputer(strategy='median')
numeric_features = ['cap-diameter', 'stem-height', 'stem-width']
X[numeric_features] = numerical_imputer.fit_transform(X[numeric_features])
```

```python
# For categorical features, impute with most frequent value
categorical_imputer = SimpleImputer(strategy='most_frequent')
categorical_features = X.select_dtypes(include=['object']).columns
X[categorical_features] = categorical_imputer.fit_transform(X[categorical_features])
```

```python
X.isnull().sum()
```

```python
# One-hot encode categorical features (creates binary columns for each category)
X_encoded = pd.get_dummies(X, columns=categorical_features, drop_first=True)
```

```python

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y['class'])
```

```python
# 4. Scale numeric features
scaler = StandardScaler()
X[numeric_features] = scaler.fit_transform(X[numeric_features])
```

```python
X_encoded
```

```python
# Align y with deduplicated X
y_aligned = y.loc[X.index].reset_index(drop=True)
y_encoded_aligned = le.transform(y_aligned['class'])

# Split the data into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_encoded_aligned, 
    test_size=0.2, 
    random_state=42, 
    stratify=y_encoded_aligned  # Maintain class balance in splits
)
```

```python
numeric_cols_in_encoded = []
for feature in numeric_features:
    numeric_cols_in_encoded.extend([col for col in X_encoded.columns if col.startswith(feature)])
```

```python
# Scale the numerical features
scaler = StandardScaler()
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

X_train_scaled[numeric_cols_in_encoded] = scaler.fit_transform(X_train[numeric_cols_in_encoded])
X_test_scaled[numeric_cols_in_encoded] = scaler.transform(X_test[numeric_cols_in_encoded])
```

```python
%pip install --upgrade scikit-learn imbalanced-learn

from imblearn.over_sampling import SMOTE

# Check if we have significant class imbalance
class_counts = np.bincount(y_train)
imbalance_ratio = class_counts[0] / class_counts[1]

# Apply SMOTE if imbalance is significant (typically if ratio > 1.5)
if imbalance_ratio > 1.5 or imbalance_ratio < 0.67:
    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train_scaled, y_train)
else:
    X_train_resampled, y_train_resampled = X_train_scaled, y_train
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

#1. Use ANOVA F-test to select top 15 features
anova_selector = SelectKBest(score_func=f_classif, k=15)
X_train_anova = anova_selector.fit_transform(X_train, y_train)
anova_features = X_train.columns[anova_selector.get_support()]
```

```python
from sklearn.ensemble import RandomForestClassifier

# 2. Use Random Forest feature importance to confirm selection
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
# Get feature importances
importances = rf.feature_importances_
feature_importance_df = pd.DataFrame({
    'feature': X_train.columns,
    'importance': importances
}).sort_values('importance', ascending=False)
```

```python
top_rf_features = feature_importance_df.head(15)['feature'].values
final_features = list(set(anova_features) & set(top_rf_features))
```

```python
if len(final_features) < 10:
    # Add top features from ANOVA
    for feature in anova_features:
        if feature not in final_features and len(final_features) < 15:
            final_features.append(feature)
    
    # Add top features from Random Forest
    for feature in top_rf_features:
        if feature not in final_features and len(final_features) < 15:
            final_features.append(feature)
```

```python
#  Prepare the final datasets
X_train_final = X_train[final_features]
X_test_final = X_test[final_features]
```

```python

# Initialize and train the Random Forest model
print("Training Random Forest Classifier...")
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1  # Use all available cores for faster training
)
```

```python
# Train the model
rf_model.fit(X_train_final, y_train)

```

```python
# Make predictions
y_pred = rf_model.predict(X_test_final)
y_prob = rf_model.predict_proba(X_test_final)[:, 1]

```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_prob)

```

```python
from sklearn.metrics import classification_report

# Generate classification report
report = classification_report(y_test, y_pred, output_dict=True)
precision = report['weighted avg']['precision']
recall = report['weighted avg']['recall']
f1 = report['weighted avg']['f1-score']
```

```python
from sklearn.model_selection import cross_val_score

# Perform cross-validation
cv_scores = cross_val_score(rf_model, X_train_final, y_train, cv=5, scoring='accuracy')

# Display results
print("\nModel Evaluation Results:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"AUC: {auc:.4f}")
print(f"Cross-validation Accuracy: {cv_scores.mean():.4f} (±{cv_scores.std():.4f})")
```

```python
from sklearn.metrics import confusion_matrix

# Create confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Edible', 'Poisonous'],
            yticklabels=['Edible', 'Poisonous'])
plt.title('Confusion Matrix - Random Forest')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```
