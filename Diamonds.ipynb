3: Scaling:
1. To equalize the scale of all variables:
When variables are between different ranges (for example, a year with high values and a price with low values), if we do not scale, one variable may outweigh the other in the model's internal calculations. This is especially true for distance-based algorithms.

2. Accelerating and strengthening the gradient descent process:
Models that rely on optimization via gradient descent (such as linear regression and neural networks) require data at the same scale to converge faster and utilize a good learning rate without problems. Without scaling, the update step may be slow or erratic.

3. Improving model performance and accuracy:
Practical experiments demonstrate that the type of scaling used has a significant impact on model performance, and choosing the wrong method is worse than not using scaling at all in some cases.
